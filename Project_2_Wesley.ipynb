# Import and Sanity Check

```python
#Import state_sales.csv file
import pandas as pd
import numpy as np

df = pd.read_csv('state_sales.csv')

df.head(10)
```

```python
cal_wage = pd.read_csv('cal_min_wage.csv')
cal_wage.set_index('Year', inplace=True)
#turn minimum wage into float
cal_wage['Minimum Wage'] = cal_wage['Minimum Wage'].str.replace('$', '')
cal_wage['Minimum Wage'] = cal_wage['Minimum Wage'].astype(float)


nyc_wage = pd.read_csv('NYC_mini_wage.csv')
nyc_wage.set_index('Year', inplace=True)
#rename wage column to minimum wage
nyc_wage.rename(columns={'Wage':'Minimum Wage'}, inplace=True)
#turn minimum Minimum wage into float
nyc_wage['Minimum Wage'] = nyc_wage['Minimum Wage'].str.replace('$', '')
nyc_wage['Minimum Wage'] = nyc_wage['Minimum Wage'].astype(float)


florida_wage = pd.read_csv('Florida_min_wage.csv')
florida_wage.set_index('Year', inplace=True)
florida_wage['Minimum Wage'] = florida_wage['Minimum Wage'].str.replace('$', '')
florida_wage['Minimum Wage'] = florida_wage['Minimum Wage'].astype(float)

fed_wage = pd.read_csv('Federal_mini_wage.csv')
fed_wage.set_index('Year', inplace=True)
fed_wage['Minimum Wage'] = fed_wage['Minimum Wage'].str.replace('$', '')
fed_wage['Minimum Wage'] = fed_wage['Minimum Wage'].astype(float)

print(cal_wage.dtypes)
```

# Initial Exploration

```python
#percentage of change in minimum wage in California year over year and store it in a new column
cal_wage['Minimum Wage Percentage'] = cal_wage['Minimum Wage'].replace(0, np.NaN)
cal_wage['Minimum Wage Percentage'] = cal_wage['Minimum Wage'].pct_change()*100

#percentage of change in minimum wage in NYC year over year and store it in a new column
nyc_wage['Minimum Wage Percentage'] = nyc_wage['Minimum Wage'].replace(0, np.NaN)
nyc_wage['Minimum Wage Percentage'] = nyc_wage['Minimum Wage'].pct_change()*100

#percentage of change in minimum wage in Florida year over year and store it in a new column
florida_wage['Minimum Wage Percentage'] = florida_wage['Minimum Wage'].replace(0, np.NaN)
florida_wage['Minimum Wage Percentage'] = florida_wage['Minimum Wage'].pct_change()*100

#percentage of change in minimum wage in Federal year over year and store it in a new column
fed_wage['Minimum Wage Percentage'] = fed_wage['Minimum Wage'].replace(0, np.NaN)
fed_wage['Minimum Wage Percentage'] = fed_wage['Minimum Wage'].pct_change()*100
```

```python
# rename and drop some colums
update_df = df.drop('FAH sales million nominal U.S. dollars with taxes and tips', axis=1)
update_df = update_df.drop('FAFH sales million nominal U.S. dollars with taxes and tips', axis=1)
update_df = update_df.drop('FAH sales million constant 1988 U.S. dollars with taxes and tips', axis=1)
update_df = update_df.drop('FAFH sales million constant 1988 U.S. dollars with taxes and tips', axis=1)
```

```python
update_df.rename(columns={'Total sales million nominal U.S. dollars with taxes and tips': 'Total sales in millions'}, inplace=True)
update_df.rename(columns={'Total sales million constant 1988 U.S. dollars with taxes and tips': 'Total sales in millions constant 1988'}, inplace=True)
print(update_df.columns)

#Turn specific columns into float
update_df['Total sales in millions'] = update_df['Total sales in millions'].str.replace(',', '').astype(float)
update_df['Total sales in millions constant 1988'] = update_df['Total sales in millions constant 1988'].str.replace(',', '').astype(float)
print(update_df.dtypes)
```

```python
#Group the df by year and find the total sales in millions
grouped = update_df.groupby('Year')['Total sales in millions'].sum()
#name the new column Total sales in millions for the year
grouped = grouped.reset_index(name='Total sales in millions for the year')
#set the year as the index
grouped = grouped.set_index('Year')
#export to xslx file
grouped.to_excel('total_sales_in_millions.xlsx')


#create a bar graph of the total sales in millions for each year
import matplotlib.pyplot as plt
grouped.plot(kind='bar', figsize=(10, 5))
plt.ylabel('Total sales in millions')
plt.title('Total sales in millions for each year')
plt.show()
```

```python
df_US_avg = update_df.groupby('Year').agg({'Total sales in millions': 'mean'})
df_US_avg.rename(columns={'Total sales in millions': 'Average total sales in millions'}, inplace=True)
df_US_avg.head(26)
```

```python
# Find the percentage growth of total sales in millions by year for each year using the grouped df
grouped['Total sales growth'] = grouped['Total sales in millions for the year'].pct_change()
#Format the percentages to look like percentages
grouped['Total sales growth percentage'] = grouped['Total sales growth'].map('{:.2%}'.format)

grouped.head(26)
```

```python
#Find the percentage of change year oer year for each state
update_df['Total sales in percentage'] = update_df.groupby('State')['Total sales in millions'].pct_change()
update_df['Total percentage in Decimal'] = update_df['Total sales in percentage']
update_df['Total sales in percentage'] = update_df['Total sales in percentage'].map('{:.2%}'.format)
update_df.head(30)
```

```python
#Group the data by state and find the max value of the total sales in millions by state
max_sales = update_df.groupby('State')['Total sales in millions'].max()
#sort the values in descending order
max_sales = max_sales.sort_values(ascending=False)
#export to xlsx file
max_sales.to_excel('max_sales.xlsx')

max_sales.head(53)
```
